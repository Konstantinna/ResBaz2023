```python
import micropip
await micropip.install("altair")
import altair as alt
```

```python
# Altair in Lite
dfbiggest = df[df["fans"] > 1000]
alt.Chart(dfbiggest).mark_bar().encode(x = 'origin',y = 'fans')
```

```python
# Plotly Example in Classic Notebooks
#fig = px.scatter(horrordf, x="Review Rating", y="Movie Run Time in", color="Release Country", hover_data=['Title'])
#fig.show()
```

```python
# Scatterplot 
alt.Chart(horrordf).mark_circle(size=60).encode(
    x='Movie Run Time in',
    y='Review Rating',
    color='Release Country',
    tooltip=['Title']
).interactive()
```

```python
# Boxplot for the all movies
alt.Chart(horrordf).mark_boxplot(extent='min-max').encode(
    x='Review Rating',
    y='Release Country',
    color='Release Country',
    tooltip=['Title']
).interactive()
```

```python
tophorror = horrordf[horrordf["Review Rating"] > 7]
```

```python
# Boxplot for the top rated ones
alt.Chart(tophorror).mark_boxplot(extent='min-max').encode(
    x='Review Rating',
    y='Release Country',
    color='Release Country',
    tooltip=['Title']
).interactive()
```

```python
# Budget by Review Rating
alt.Chart(horrordf).mark_circle(size=60).encode(
    x='Budget',
    y='Review Rating',
    color='Release Country',
    tooltip=['Title']
).interactive()
```
